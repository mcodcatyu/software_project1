```python
import LebwohlLasher_vectorization as lv
import LebwohlLasher  as ll
import matplotlib.pyplot as plt
import time
test_inputs = [
    ("LL", 10, 50, 0.5, 0),
    ("LL", 100, 50, 0.5, 0),
    ("LL", 1000, 50, 0.5, 0),
    ("LL", 2000, 50, 0.5, 0),
]
values1 = []
results1 = []
values2 = []
results2 = []
```

```python


for program, nsteps, nmax, temp, pflag in test_inputs:
    values1.append(nsteps)

    start = time.perf_counter()
    lv.main(program, nsteps, nmax, temp, pflag )
    end = time.perf_counter()

    duration = end - start

    results1.append(duration)


for program, nsteps, nmax, temp, pflag in test_inputs:
    values2.append(nsteps)

    start = time.perf_counter()
    ll.main(program, nsteps, nmax, temp, pflag )
    end = time.perf_counter()

    duration = end - start

    results2.append(duration)
```

```python
print(values1,
results1,
values2,
results2)
```

```python
fig, ax = plt.subplots(figsize = (10, 6))
plt.plot(values1, results1, label = 'vectorization', marker = 'o', linestyle ='-')
plt.plot(values2, results2,  label = 'without vectorization', marker = 'o', linestyle ='-')
plt.title('vectorization vs original')
plt.xlabel('step')
plt.ylabel('time(s)')
plt.legend()
plt.show()
```
